```python
#Dependencies and Setup
import pandas as pd
import numpy as np

#Load data from purchase_data.csv
gamers_file = "Resources/purchase_data.csv"

#Read purchase_data.csv and store into Pandas dataframe
gamers_data_df = pd.read_csv(gamers_file)

gamers_data_df.head()
```

## Player Count

* Display the total number of players


```python
#Calculate total number of unique players
player_count = len(gamers_data_df.groupby(["SN"]))

#Display total # of players
total_players = pd.DataFrame({"Total Players": [player_count]})
total_players
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
#Calculate number of unique items under "Item Name" in the gamers_data_df DataFrame
items_count = len(gamers_data_df["Item ID"].unique())

#Calculate average item price under "Price" in the gamers_data_df DataFrame
avg_item_price = "${:,.2f}".format(gamers_data_df["Price"].mean())

#Calculate total number of items purchased under "Item Name" in the gamers_date_df DataFrame
items_purchased = len(gamers_data_df["Item Name"])

#Calculate total revenue under "Price" in the gamers_data_df DataFrame
total_revenue = "${:,.2f}".format(sum(gamers_data_df["Price"]))

#def format(x):
 #   return "${:.2f}K".format

purchasing_analysis_table = pd.DataFrame({"Number of Unique Items": [items_count], 
                                          "Average Price": [avg_item_price], 
                                          "Number of Purchases": [items_purchased], "Total Revenue": [total_revenue]})
purchasing_analysis_table
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python

gender_df = pd.DataFrame(gamers_data_df.groupby(['SN', 'Gender']).size().reset_index().rename(columns={0:'Frequency'}))
gender_counts = gender_df["Gender"].value_counts()

males_total = gender_counts.loc["Male"]
males_percentage = "{:,.2f}%".format(males_total / player_count * 100) 
males_df = pd.DataFrame({"Total Count": [males_total], 
                                          "Percentage of Players": [males_percentage]})
                         
females_total = gender_counts.loc["Female"]
females_percentage = "{:,.2f}%".format(females_total / player_count * 100) 
females_df = pd.DataFrame({"Total Count": [females_total], 
                                          "Percentage of Players": [females_percentage]})

other_total = gender_counts.loc["Other / Non-Disclosed"]
other_percentage = "{:,.2f}%".format(other_total / player_count * 100) 
other_df = pd.DataFrame({"Total Count": [other_total], 
                                          "Percentage of Players": [other_percentage]})

gender_df = [males_df, females_df, other_df]
gender_demo_df = pd.concat(gender_df)
gender_demo_df.index = ["Male", "Female" , "Other/Non-Disclosed"]

gender_demo_df
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
grpBy=gamers_data_df.groupby("Gender")
grpBy.count()
gamers_data_df.SN.nunique()

pd.DataFrame({
    "Purchase Count":grpBy.Age.count(),
    "Average Purchase Price":grpBy.Price.mean(),
    "Total Purchase Value":grpBy.Price.sum(),
    "Avg Total Purchase per Person":grpBy.Price.sum() / gender_counts
},grpBy.count().index)
```

```python
bins = [0, 9, 14, 19, 24, 29, 34, 39, 99]
group_labels = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

player_count1 = (gamers_data_df[['SN', "Age"]].groupby("SN").max())
player_count1


player_count1["Age Group"] = pd.cut(player_count1["Age"], bins, labels=group_labels)                            
player_count1                                 

age_df = player_count1[['Age', 'Age Group']].groupby("Age Group").count().rename(columns = {"Age": "Total Count"})

age_df["Percent of Players"] = age_df["Total Count"] /player_count * 100
age_df

#age_df["Percent of Players"] = avg_df["Perecnt of Players"].map("{:.2f}%".format)
   


                           
                           
```
